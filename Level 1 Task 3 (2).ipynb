**2: Explore the Distribution of categorical variables**

```python
# Explore the distribution of categorical variables 
categorical_cols = file.select_dtypes(include=['object']).columns
```

```python
for col in categorical_cols:
    print(f"Distribution of {col}")
    
print(file[col].value_counts())
```

**3: Identify the Top Cuisine and Cities with the Highest Number of Restaurants**

```python
# Identify the top cuisine with the highest number of restaurants 
top_cuisine = file['Cuisines'].value_counts().head(10)
print("\nTop 10 Cuisines:")
print(top_cuisine)

# Identify the top cities with the highest number of restaurants 
top_cities = file['City'].value_counts().head(10)
print("\nTop 10 Cities:")
print(top_cities) 
```

**4: Visualize top cuisines and top cities**

```python
#Visualize top 10 cuisines 
plt.figure(figsize=(16,8))
sns.barplot(x=top_cuisine.index, y=top_cuisine.values)
plt.xlabel('Cuisines')
plt.ylabel('Count')
plt.title('Top 10 Cuisines')
plt.show()

```

```python
#Visualize top 1o cities
plt.figure(figsize=(12,6))
sns.barplot(x=top_cities.index,y= top_cities.values)
plt.ylabel('Counts')
plt.xlabel('Cities')
plt.title("Top 10 cities")
plt.show()
```

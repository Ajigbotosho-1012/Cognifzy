**4: Analyze the Distribution of the Target Variable**

**5: Check for class imbalance**

```python
# Get the target column
target = file['Aggregate rating']

# calculate the class counts
class_counts = {}
for label in target:
    if label in class_counts:
        class_counts[label] += 1
    else:
        class_counts[label] = 1
        
# calculate imbalance ratio
imbalance_ratio = max(class_counts.values()) / min(class_counts.values())

print('\nClass counts:', class_counts)
print('\nImbalance Ration:', imbalance_ratio)
```

```python
print(len(class_counts))
```

**6: Visualize the class imbalance**

```python
# Get the x and y values
labels = list(class_counts.keys())
counts = list(class_counts.values())
```

```python
print(labels)
print(counts)
```

```python
# Visualize using barplot 
plt.figure(figsize=(10,8))
sns.barplot(x=labels, y=counts)
plt.xlabel('Class Label')
plt.ylabel('Count')
plt.title('Class Imbalance Visualization')
plt.ylim(0,1000,50)
plt.show()
```
